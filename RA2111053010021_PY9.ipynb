```python
#Angad Singh Hoonjan
#RA2111053010021
import pandas as pd
df1 = pd.DataFrame({'A':[10,20,30], 'B':[40,50,60]})
print('\nData Frame 1: \n',df1)
df2 = pd.DataFrame({'A':[40,50,60], 'B':[75,85,95]})
print('\nData Frame 2: \n',df2)
df3 = df1.append(df2)
print('\nData Frame appended: \n',df3)
df4 = df1.append(df2,ignore_index = True)
print('\nData Frame appended without index: \n',df4)
df5 = pd.concat([df1,df2], axis=1)
print('\nConcatenated Data Frame: \n',df5)
df6 = pd.concat([df1,df2], keys=['UK','DE'])
print('\nConcatenated Data Frame with Keys: \n',df6)
df7 = df6.loc["DE"]
print('\nData Frame with particular key/index: \n',df7)
```

```python
#Angad Singh Hoonjan
#RA2111053010021
import pandas as pd
d1 = {'Name':['Arjit','Meghna','Sunny'], 
      'Country':['India','India','USA'], 
      'Role':['CEO','Sales Manager','Director']}
df1 = pd.DataFrame(d1)
print('DataFrame 1:\n', df1)
d2 ={'EmpID': [1, 2, 3],'Name':['Vishal','Arjit','Roopam']}
df2 = pd.DataFrame(d2)
print('\nDataFrame 2:\n',df2)

# default merge-inner join
df_merged = df1.merge(df2)
print('\nResult with default merge(inner join) : \n', df_merged)

# Merging DataFrames with Left, Right, and Outer Join
dfl= df1.merge(df2, how='left')
dfr= df1.merge(df2, how='right')
dfo= df1.merge(df2, how='outer')
print('\nResult Left Join: \n', dfl)
print('\nResult Right Join: \n', dfr)
print('\nResult Outer Join: \n', dfo)

# Merging DataFrame on Specific Columns
dfn= df1.merge(df2, on='Name')
print('\nResult merge on name column: \n',dfn)

# Specify Left and Right Columns for Merging DataFrame Objects
nd1 = {'Name': ['Arjit','Meghna','Sunny'], 
       'EmpID1': [1, 2, 3],
       'Country': ['India','India','USA'], 
       'Role': ['CEO','Sales Manager','Director']}
ndf1 = pd.DataFrame(nd1)
nd2 ={'EmpID2': [1, 2, 3],
      'Name': ['Vishal','Arjit','Roopam']}
ndf2 = pd.DataFrame(nd2)
print('\nResult merge on left and right column: \n',ndf1.merge(ndf2,left_on='EmpID1',right_on='EmpID2'))

# Using Index as the Join Keys for Merging DataFrames
dfi = df1.merge(df2, left_index=True, right_index=True)
print('\nResult Index Merge: \n', dfi)
```
