```python
#Angad Singh Hoonjan
#RA2111053010021

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate 20 random data points
np.random.seed(42)
data = {'X': np.random.rand(20), 'Y': np.random.rand(20)}

# Create a Pandas DataFrame from the random data
df = pd.DataFrame(data)

# Scatter plot
plt.figure(figsize=(4,4))
plt.scatter(df['X'], df['Y'], color='RED', label='Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Angad Singh Hoonjan
#RA2111053010021

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate 20 random data points
np.random.seed(42)
data = {'X': np.random.rand(20)}

# Create a Pandas DataFrame from the random data
df = pd.DataFrame(data)

# KDE plot
plt.figure(figsize=(4,4))
sns.kdeplot(df['X'], color='green', label='KDE X')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('KDE Plot')
plt.legend()

plt.show()
```
