```python
#Sangeetha Nair
#RA2111053010026
import matplotlib.pyplot as plt
import numpy as np

#Generate x value from 0 to 2*pi
x = np.linspace(0,4*np.pi,100)

#Calculate y values using sine function
y = np.sin(x)

#Create a line plot
plt.plot(x,y)

#Add labels and title
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Function')
plt.axhline(y=0, color='black', linestyle='-.')
```

```python
#Sangeetha Nair
#RA2111053010026

import matplotlib.pyplot as plt

#Define the population data
male=550
female=650
#Calculate percentage
total = male+female
male_p = (male/total)*100
female_p = (female/total)*100

#Labels
labels = ['Male', 'Female']

#Size
sizes = [male_p, female_p]

#Colors
colors = ['paleturquoise','lightpink']

#Create pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=130)

plt.axis('equal')
plt.title('Population Distribution by Gender')
```
