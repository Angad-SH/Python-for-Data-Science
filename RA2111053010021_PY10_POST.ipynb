```python
#Angad Singh Hoonjan
#RA2111053010021

import matplotlib.pyplot as plt

# Create a figure with a specific size
fig = plt.figure(figsize=(2.5,2.5))

# Define the positions and sizes of the axes using add_axes
# The parameters are [left, bottom, width, height]
ax1 = fig.add_axes([0, 0, 1, 1])  # Main plot
ax2 = fig.add_axes([0.2, 0.2, 0.2, 0.2])  # Inset plot
ax3 = fig.add_axes([0.65, 0.65, 0.3, 0.3])

# Show the plot
plt.show()
```
